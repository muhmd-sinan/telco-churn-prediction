# Churn EDA — Telco Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/Telco-Customer-Churn.csv')

# convert TotalCharges to number (raw CSV stores it as text)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# convert Churn to 0/1 so we can use .mean() for churn rate
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

print(df.shape)
df.head()
```

```python
df.dtypes
```

```python
blank_charges = df[df['TotalCharges'].str.strip() == '']
print('Blank TotalCharges rows:', len(blank_charges))
blank_charges[['customerID', 'tenure', 'MonthlyCharges', 'TotalCharges']]
```

## Churn Distribution

```python
df['Churn'].value_counts().plot(kind='bar')
plt.title('Churn Distribution')
plt.ylabel('Count')
plt.show()

print('Churn rate:', df['Churn'].mean() * 100, '%')
```

## Numerical Distributions

```python
df[df['Churn'] == 0]['tenure'].hist(alpha=0.6, label='No Churn')
df[df['Churn'] == 1]['tenure'].hist(alpha=0.6, label='Churned')
plt.title('Tenure by Churn')
plt.legend()
plt.show()
```

```python
df[df['Churn'] == 0]['MonthlyCharges'].hist(alpha=0.6, label='No Churn')
df[df['Churn'] == 1]['MonthlyCharges'].hist(alpha=0.6, label='Churned')
plt.title('MonthlyCharges by Churn')
plt.legend()
plt.show()
```

```python
df[df['Churn'] == 0]['TotalCharges'].hist(alpha=0.6, label='No Churn')
df[df['Churn'] == 1]['TotalCharges'].hist(alpha=0.6, label='Churned')
plt.title('TotalCharges by Churn')
plt.legend()
plt.show()
```

## Churn by Contract Type

```python
contract_churn = df.groupby('Contract')['Churn'].mean() * 100
contract_churn.plot(kind='bar')
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate (%)')
plt.show()
```

## Churn by Internet Service

```python
inet_churn = df.groupby('InternetService')['Churn'].mean() * 100
inet_churn.plot(kind='bar')
plt.title('Churn Rate by Internet Service')
plt.ylabel('Churn Rate (%)')
plt.show()
```

## Churn by Payment Method

```python
pay_churn = df.groupby('PaymentMethod')['Churn'].mean() * 100
pay_churn.plot(kind='bar')
plt.title('Churn Rate by Payment Method')
plt.ylabel('Churn Rate (%)')
plt.show()
```

## Churn by Tenure Bucket

```python
bins = [0, 12, 24, 48, 72]
labels = ['0-12', '13-24', '25-48', '49-72']
df['tenure_bucket'] = pd.cut(df['tenure'], bins=bins, labels=labels)

tenure_churn = df.groupby('tenure_bucket')['Churn'].mean() * 100
tenure_churn.plot(kind='bar')
plt.title('Churn Rate by Tenure Bucket')
plt.ylabel('Churn Rate (%)')
plt.show()
```

## Correlation Matrix

```python
corr = df[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']].corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

## Key Findings

- 26.5% churn rate — class imbalance
- Contract type = strongest signal: Month-to-month ~42% churn, Two year ~3%
- New customers (0-12 months tenure) churn most
- Fiber optic internet = highest churn
- Electronic check = highest churn by payment method
- Higher MonthlyCharges = more churn
- 11 blank TotalCharges = new customers (tenure=0)
